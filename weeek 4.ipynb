The most basic type of plot you can make is a line chart with data defined as either Python lists or NumPy arrays. As we'll see, the parameters can be any array-like or scalar value.

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [1, 2, 3, 4, 5, 6, 7]

plt.plot(x, y);
```

This gives us a very plain line chart with x and y axis scales but no labels. The size and color of the plot line use Matplotlib's default values. Change the color by setting a `color` attribute.

```python

x = np.array([1, 2, 3, 4, 5, 6, 7])
y = np.array([1, 1, 2, 3, 5, 8, 13])

plt.plot(x, y, color='red');
```

Matplotlib makes it very easy to display plots with different line styles.

```python
y = [2, 5, 9, 13, 18]

plt.plot(y, linestyle='dotted');
```

Note that we only provided y values and Matplotlib filled in a set of default values for the x-axis.

There are many different line styles available. For example, you can display a combination of dashes and dots by specifying the line style of `'-.'`. We can also change the line width.

```python
plt.plot(y, linestyle='-.', linewidth=5);
```

You can view `matplotlib.lines.lineStyles` to see what line styles are available.

```python
matplotlib.lines.lineStyles
```

You can display multiple lines on the same plot by calling the `plot` function multiple times.

```python
y1 = [1, 2, 3, 4, 5]
y2 = [2, 5, 9, 13, 18]

plt.plot(y1)
plt.plot(y2);
```

You can also pass multiple sets of x and y data points to the `plot` function at once.

```python
x1 = [1, 2, 3, 4, 5]
y1 = [1, 2, 3, 4, 5]
x2 = [1, 2, 3, 4, 5]
y2 = [2, 5, 9, 13, 18]
x3 = [1, 2, 3, 4, 5]
y3 = [2, 4, 6, 8, 10]

plt.plot(x1, y1, x2, y2, x3, y3);
```

<b>Markers</b>
</div>

You can also control the markers used to plot each data point.

```python
y = [2, 5, 9, 13, 18]

plt.plot(y, marker='o');  # That's the lowercase letter 'o'
```

There are a lot of different markers to choose from.

```python
y1 = [1, 2, 3, 4, 5]
y2 = [2, 5, 9, 13, 18]

plt.plot(y1, marker='s')  # square marker
plt.plot(y2, marker='D'); # diamond marker
```

You can also change the marker size.

```python
y = [2, 5, 9, 13, 18]

plt.plot(y, marker='o', markersize=12);
```

We can also change the marker edge color and face color.

```python
plt.plot(y, marker='o', markersize=12, markeredgecolor='red', markerfacecolor='yellow');
```

You can see all of the different marker styles using `matplotlib.markers.MarkerStyle.markers`, or view all the symbols available at [matplotlib.markers](https://matplotlib.org/stable/api/markers_api.html).

Since Matplotlib takes any array-like or scalar value as its data parameters, we can use it to plot continuous functions (if we just give it enough data points to plot).

```python
x = np.linspace(start=-np.pi, stop=np.pi, num=256)
y = np.sin(2*x)

plt.plot(x, y);
```

We can also fill colors between two lines on the same plot. In the plot below, notice that the second set of y values is a scalar. This tells Matplotlib to fill the area between the first function and the 0 y-coordinate.

```python
plt.plot(x, y)
plt.fill_between(x, y1=y, y2=0, alpha=0.1);
```

If you look at the plot above, it can be difficult to estimae the values of the data points on each plot line. Let's add markers and grid lines to make this easier. For this first example, I'll just call `grid(True)` to display grid lines with the default settings.

```python
x = ['Jan', 'Feb', 'March', 'Apr', 'May']
y1 = [1, 2, 3, 4, 5]
y2 = [2, 5, 9, 13, 18]
y3 = [2, 4, 6, 8, 10]

plt.plot(x, y1, x, y2, x, y3, marker='o')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales')
plt.legend(labels=['Alice', 'Bob', 'Carol'], loc='upper left')
plt.grid(True);
```

```python
x = ['Jan', 'Feb', 'March', 'Apr', 'May']
y1 = [1, 2, 3, 4, 5]
y2 = [2, 5, 9, 13, 18]
y3 = [2, 4, 6, 8, 10]

plt.plot(x, y1, x, y2, x, y3, marker='o')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales')
plt.legend(labels=['Alice', 'Bob', 'Carol'], loc='upper left')
plt.grid(color='blue', alpha=0.1, linewidth=3);
```

<b>Bar plots</b>
</div>

The data for a bar plot is different from the data we've seen up until now. For line charts, we were using numeric data on both the x and y axes. For a bar plot, we need our x-axis data to be a set of labels for the bars, and the y-axis data should be a numeric value that corresponds to the height of each bar.

Let's create data for our first bar plot using the `Embarked` column from the Titanic data set.

```python
embarked = titanic_df['Embarked'].value_counts()
print(embarked.index) # these will be our bar labels
print(embarked.values) # these will be the heights of the bars
```

We create a bar plot using the [`matplotlib.pyplot.bar`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html) function.

```python
import matplotlib.pyplot as plt
plt.bar(x=embarked.index, height=embarked.values);
```

We can also specify a list of colors to apply to each bar.

```python
plt.bar(x=embarked.index, height=embarked.values, color=['blue', 'green', 'red']);
```

We can also create a horizontal bar plot using the `barh` function, but we have to specify the labels on the y-axis the values as bar widths instead of heights.

```python
plt.barh(y=embarked.index, width=embarked.values, color=['blue', 'green', 'red']);
```

<b>Stacked bar plots</b>

To create stacked bar plots we can separate the embarked data into two groups for men and women.

```python
embarked_m = titanic_df[titanic_df['Sex']=='male']['Embarked'].value_counts()
embarked_f = titanic_df[titanic_df['Sex']=='female']['Embarked'].value_counts()
print(embarked_m)
print(embarked_f)
```

In order to stack the bars, we plot one set of data normally, then we tell Matplotlib to stack the other on top using the `bottom` parameter. This specifies the y-coordinate of the bars bases.

```python
plt.bar(x=embarked_m.index, height=embarked_m.values, color='blue')
plt.bar(x=embarked_f.index, height=embarked_f.values, bottom=embarked_m.values, color='red');
```

Finally, we can use the same functions that we saw before to provide axis labels, a title, and a plot legend.

```python
plt.bar(x=embarked_m.index, height=embarked_m.values, color='blue', width=0.75)
plt.bar(x=embarked_f.index, height=embarked_f.values, bottom=embarked_m.values, color='red', width=0.75)

plt.xlabel('Port')
plt.ylabel('Passengers')
plt.title('Port of Embarkation by Sex')
plt.legend(labels=['M', 'F']);
```

<b>Histograms</b>

We can use Matplotlib's [`hist`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html) function to display a frequency distribution of some variable. Let's show the distribution of the ages of passengers on the Titanic.

```python
ages = titanic_df['Age']
plt.hist(ages);
```

Note that this illustrates the total count (y-axis) of the ages (x-axis) of the passengers separated into 10 bins. We can show a probability density of ages in each bin instead of the total count by changing to a density chart.

```python
plt.hist(ages, density=True);
```

We can also change the number of bins from the default to make the histogram more or less granular, and add titles, labels, and gridlines to make it more readable.

```python
plt.hist(ages, bins=20)

plt.xlabel('Age')
plt.ylabel('Passengers')
plt.title('Distribution of Passengers Ages')
plt.grid(True);
```

From this distribution, we can see that most of the passengers were in the age range of about 20 to 40. We can also see that there were many young children aboard, and relatively few passengers over the age of 70.

<b>Pie charts</b>

To make a pie chart, use Matplotlib's [`matplotlib.pyplot.pie`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html) function, which expects a list representing the sizes of each slice. We can use the count of male and female passengers as an example.

```python
titanic_df = pd.read_csv("titanic.csv")
x = titanic_df['Sex'].value_counts()

plt.pie(x);
```

Without labels, we can't really understand what even the most basic pie chart means.

```python
plt.pie(x, labels=['Male', 'Female']);
```

That's better, but what if we want the first slice to start at the 12 o'clock position, which is standard? The default seems to be starting at 3 o'clock. We can do that by specifying a `startangle`.

```python
plt.pie(x, labels=['Male', 'Female'], startangle=90);
```

You can separate one or more slices from the rest of the chart using the `explode` parameter.

```python
plt.pie(x, labels=['Male', 'Female'], startangle=90, explode=[0.1, 0]);
```

Finally, we can display the percentage that each slice represents using the `autopct` parameter by setting it to a format specifier.

```python
plt.pie(x, labels=['Male', 'Female'], startangle=90, autopct='%.1f%%');
```

<b>Scatter plots</b>

To create scatter plots, use Matplotlib's [`matplotlib.pyplot.scatter`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html) function. It expects two lists of values representing x and y coordinates. Let's use the Age and Fare columns from the Titanic data set.

```python
age = titanic_df['Age']
fare = titanic_df['Fare']

plt.scatter(x=age, y=fare);
```

There doesn't seem to be any correlation between the age of passengers aboard the Titanic and the fare they paid for passage.

We can also plot multiple datasets on the same chart.

```python
age_m = titanic_df[titanic_df['Sex'] == 'male']['Age']
fare_m = titanic_df[titanic_df['Sex'] == 'male']['Fare']
age_f = titanic_df[titanic_df['Sex'] == 'female']['Age']
fare_f = titanic_df[titanic_df['Sex'] == 'female']['Fare']

plt.scatter(x=age_m, y=fare_m)
plt.scatter(x=age_f, y=fare_f)
plt.legend(['male', 'female']);
```

Since we have another numeric data column (Pclass), we can use either color or marker size to represent a third axis.

```python
age = titanic_df['Age']
fare = titanic_df['Fare']
pclass = titanic_df['Pclass']
fig = plt.figure(figsize=(12, 6))
plt.scatter(x=age, y=fare, c=pclass, cmap='RdYlGn')
plt.colorbar(ticks=[1, 2, 3])
plt.xlabel('Age')
plt.ylabel('Fare');
```

Note that this time I specificed a `cmap` (colormap) and colorbar instead of using a legend. This tells us that their is a relationship between fare and class. Most of the 1st class passengers paid the higher fares.

Let's try showing the class using the size of each marker.

```python
fig = plt.figure(figsize=(12, 6))
plt.scatter(x=age, y=fare, s=(pclass*35))
plt.xlabel('Age')
plt.ylabel('Fare');
```

If you look closely, you can see that the markers near the bottom of the plot are larger, indicating those passengers were in a higher class then passengers plotted higher up. It's much easier to see this distinction using colors, at least in this case.

<b>Box plots</b>

You can create box plots (box-and-whisker plots) in Matplotlab using the [`matplotlib.pyplot.boxplot`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.boxplot.html) function. Box plots are important in statistical analysis because they let you see the distrubution of a data set at a glance, including outliers. Let's look at the age of Titanic passengers again, this time in a box plot.

```python
age = titanic_df['Age'].dropna() # Drop missing values before plotting or the boxplot will be empty
fig = plt.figure(figsize=(6, 6))
plt.boxplot(age, labels=['Age']);
```

We can pick a few key datapoints off the boxplot at a glance. The median age is just under 30, with half of the ages in the 20 to 40 range. The entire distribution of ages is from 0 to 80 years, with a few data points at the high end of the scale considered outliers.

<b>3D plots</b>
</div>

Earlier we showed how color and marker size can be used as a third dimension in Matplotlib, but you can also plot that data on a third axis by setting the projection to `3d` when calling the `axes` function. That function returns an axis object that we use to create a scatterplot.

```python
age = titanic_df['Age']
fare = titanic_df['Fare']
pclass = titanic_df['Pclass']

fig = plt.figure(figsize=(9, 9))
ax = plt.axes(projection='3d')
ax.scatter(xs=age, ys=fare, zs=pclass)
ax.set_xlabel('Age')
ax.set_ylabel('Fare')
ax.set_zlabel('Class');
```

Like on the 2D chart, we can see that the Fare doesn't vary much by Age in 2nd and 3rd class on Titanic, but it varies quite a bit in 1st class.

Even with classes separated as they are, it's still difficult to see which data points belong to which class when they're all the same color. Just like with a 2D scatter plot, we can set the color for each class to make it easier to tell them apart.

```python
fig = plt.figure(figsize=(9, 9))
ax = plt.axes(projection='3d')
ax.scatter(xs=age, ys=fare, zs=pclass, c=pclass, cmap='brg')
ax.set_xlabel('Age')
ax.set_ylabel('Fare')
ax.set_zlabel('Class');
```

<b>Subplots</b>
</div>

Many times you'll need to plot multiple charts in one figure. Matplotlib allows you to do this with the [`matplotlib.pyplot.subplots`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html) function, where you specify the number of rows and columns, and it returns a grid of axes. This will make more sense with an example. Let's draw two pie charts from the Titanic dataset in one grid, side-by-side. 

```python
s_vals = titanic_df['Sex'].value_counts()
e_vals = titanic_df['Embarked'].value_counts()
fig, axes = plt.subplots(1, 2) 
fig.set_size_inches(8, 4) 
axes[0].pie(s_vals, labels=['Male', 'Female'], startangle=90, autopct='%.1f%%')
axes[0].set_title('Sex')
axes[1].pie(e_vals, labels=['S', 'C', 'Q'], startangle=90, explode=[0, 0.1, 0])
axes[1].set_title('Embarked');
```

You don't have to create the same kind of chart in each grid space.

```python
fig, axes = plt.subplots(1, 3) 
fig.set_size_inches(12, 4) 
axes[0].pie(s_vals, labels=['Male', 'Female'], startangle=90, autopct='%.1f%%')
axes[0].set_title('Sex')
age = titanic_df['Age'].dropna() 
axes[1].boxplot(age, labels=['Age']);

age = titanic_df['Age']
fare = titanic_df['Fare']
axes[2].scatter(x=age, y=fare);
```

<b>Text and Annotations</b>
</div>

Matplotlib allows you to add text and annotations to a plot so that you can label features of your data that you'd like to draw attention to. Let's go back to our earlier example of plotting a function, but this time we'll add a label saying what the function is.

```python
x = np.linspace(start=-np.pi, stop=np.pi, num=256)
y = np.sin(2*x)

fig = plt.figure(figsize=(8, 5))
plt.plot(x, y)
plt.text(-1.65, 0.5, 'y = sin(2x)', fontsize=14);
```

We could also plot and annotate a specific point.

```python
fig = plt.figure(figsize=(8, 5))
plt.plot(x, y, color='gray')

plt.plot([0], [0], 'o') 

arrowprops = dict(facecolor='black', shrink=0.1, width=2)
plt.annotate('origin point',        
             xy=(0, 0),            
             xytext=(-1.65, 0.5),   
             arrowprops=arrowprops, 
             fontsize=12            
            );
```
